```python
for i in os.listdir(dir_path):
  
    img = image.load_img(dir_path+'//'+i,target_size=(450,500))
    hlsimage = cv2.cvtColor(np.array(img),cv2.COLOR_RGB2HLS) 
    flame = hlsimage[:,:,1]
    
 
    

    for x in range(0,450):
        for y in range(0,500):
             if flame[x,y]<190 or flame[x,y]<230:
                    flame[x,y]=0
    plt.imshow(flame)  
    plt.show()
```

```python
for i in os.listdir(dir_path):
  
    img = image.load_img(dir_path+'//'+i,target_size=(450,500))
    hlsimage = cv2.cvtColor(np.array(img),cv2.COLOR_RGB2HLS) 
    leakage = hlsimage[:,:,1]
    for x in range(0,450):
        for y in range(0,500):
            
             if  leakage[x,y]<190 or  leakage[x,y]<230:
                    leakage[x,y]=0
    blur = cv2.GaussianBlur( leakage,(3,3),0)
    canny = cv2.Canny(blur,200, 255)
    plt.imshow(canny)
    plt.show()
```

```python
for i in os.listdir(dir_path):
  
    img = image.load_img(dir_path+'//'+i,target_size=(450,500))
    hlsimage = cv2.cvtColor(np.array(img),cv2.COLOR_RGB2HLS) 
    leakage = hlsimage[:,:,1]
    for x in range(0,450):
        for y in range(0,500):
             if  leakage[x,y]<190 or leakage[x,y]<230:
                    leakage[x,y]=0
    leakagearea = 0
    for y in range(0, 450):
        x = 0
        for x in range(0,400):
            #find the minimum Y
            if  leakage[x,y]>0:
                 leakagearea += 1
            x += 1
        y += 1  
    print(' leakage Area ', leakagearea)
    img= np.array(img)
    if leakagearea>1000:
             font = cv2.FONT_HERSHEY_SIMPLEX
             cv2.putText(img,"leakage_detected",(50,50),font,1,(255,255,255),3,cv2.LINE_AA)
    plt.imshow(img)
    plt.show()

           
    
          
    
          
```
